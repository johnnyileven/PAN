### Evolution

$$\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}$$

### $$\rho_W^N(t) = \frac{1}{N} \bigg(\ket{2^0}\bra{2^0} + \ket{2^1}\bra{2^1} + .. \ket{2^{N-1}}\bra{2^{N-1}} + \sum_{(i\neq j)=0}^{N-1} \ket{2^{i}}\bra{2^{j}} e^{-1/2(\chi_{ii} + \chi_{jj} - 2\chi_{ij})}  \bigg)   $$

###  $$\rho_{GHZ}^N(t) = \frac{1}{2} \bigg(\ket{0}\bra{0} + \ket{2^N-1}\bra{2^N-1} + \ket{0}\bra{2^N-1} e^{-1/2\sum_i\chi_{ii}} +  \ket{2^N-1}\bra{0} e^{-1/2\sum_i\chi_{ii}} \bigg)   $$

Where:

$$\chi_{ij} = \int {dt}_1 \int {dt}_2 <\xi_i(t_1) \xi_j(t_2)>$$
